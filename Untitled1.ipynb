```python
import pandas as pd
df = pd.read_csv('exam.csv.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
```

```python
df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
```

```python
df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()
```

```python
df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='gender', y='math score', data=df)
plt.title('Average Math Score by Gender')
plt.show()
```

```python
sns.barplot(x='gender', y='reading score', data=df)
plt.title('Average Reading Score by Gender')
plt.show()
```

```python
sns.barplot(x='test preparation course', y='math score', data=df)
plt.title('Math Score: Completed vs No Test Prep')
plt.show()
```

Findings:

Male students scored higher in math on average, while female students scored higher in reading and writing.
Students who completed the test preparation course scored higher across all subjects.
